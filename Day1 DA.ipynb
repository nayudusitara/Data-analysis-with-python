```python
#list comprehension
#newlist=[output for]
l1=[12,143,14,45,34,19,98,56,42,86]
l2=[i*i for i in l1 if i>50]
print(l2)
```

```python
#list comprehension
#newlist=[(st of if) if (cond) else (st of else) for]
l1=[12,143,14,45,34,19,98,56,42,86]
l2=[i*i  if i>50 else "X" for i in l1]
print(l2)
```
