```python
A=np.array([2,54,36,23]).reshape(2,2)
B=np.array([73,84,90,45]).reshape(2,2)
print(A)
print("---------------------")
print(B)
print("--------------------")
c=A.dot(B)#also we can use @ symbol for mat multipilication
print(c)
print("-----------------------")
print(A.max(axis=0))#0 indicate coloumns
print("------------------")
print(A.max(axis=1))#1 indicates row wise
print("-----------------------")
print(B.min(axis=0))#0 indicate coloumns
print("------------------")
print(B.min(axis=1))#1 indicates row wise
```

```python
#joining arrays!
A=np.array([2,54,36,23]).reshape(2,2)
B=np.array([73,84,90,45]).reshape(2,2)
print(A)
print("---------------------")
print(B)
print("--------------------------")
print("After vertically stacking the array!")
print(np.vstack((A,B)))
print("---------------------")
print("After Horizontally stacking the array!")
print(np.hstack((A,B)))
print(np.stack((A,B),axis=0))
print("------------------")#to print along with group(3d)

```

```python
A=np.arange(30).reshape(2,3,5)
print(A)
print("after stack")

print(np.dstack(A))#no.of rows-->no.of grps
#no.of col in every grp-->no.of rows in every grps
```

```python
A=np.arange(50).reshape(5,10)
print(A)
print("after splitting")
print(np.vsplit(A,5))#it splits vertically as 5 parts
print("---------------")
print(np.vsplit(A,(1,3)))#it splits vertically upto 1 one array and (2,3)as one array that means it splits upto 3 and remaing will under go as another array...
print("-----------")
print(np.hsplit(A,5))
print("---------------")
print(np.hsplit(A,(1,3)))
```
