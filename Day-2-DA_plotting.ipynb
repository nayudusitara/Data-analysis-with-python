```python
from matplotlib import pyplot as plt
```

```python
#an India v/s aus match is on live.IND has given 298 as target for AUS.the runs for every 5overs is given as follows
#IND=[25,51,84,131,160,189,220,250,267,297]
#AUShave compeleted 25 overs
#AUS=[15,41,94,110,151],plot overs v/s runs for both the countries with IND has blue line,
#AUS has yellow line with require labels and legend,the marker for india is flagand marker for AUS is plus.,linewidth=2
#create overs from arange functions,create runs as array.
import numpy as np
IND=np.array([25,51,84,131,160,189,220,250,267,297])
AUS=np.array([15,41,94,110,151])
overs1=np.arange(5,51,5)
overs2=np.arange(5,26,5)
plt.plot(overs1,IND,label="INDIA",color="b",linewidth=2,marker=">")
plt.plot(overs2,AUS,label="AUSTRALIA",color="y",linewidth=2,marker="P")
plt.legend()
plt.xlabel("overs")
plt.ylabel("runs")
plt.title("INDIA V/S AUSTRALIA")
plt.grid()

```
