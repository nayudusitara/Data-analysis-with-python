```python
 #printing tuple
 tup=(1,34,"swathi","situ","appu",23.6,1)
 print(tup)
 print(len(tup))
 print(tup.count(1))
 print(tup.index("appu"))
 tup1=(23,45,12,97,34,89)
 print(sorted(tup1))
 print(max(tup1))
 print(min(tup1))
 tup2=("apple","kiwi","black berry","strawberry","mango","custurdapple")
 print(sorted(tup2))
 print(max(tup2))
 print(min(tup2))
```

```python
#tuple concantination
tup1=(1,2,3)
tup2=("apple","kiwi","black berry","strawberry","mango","custurdapple")
tup3=tup1+tup2
print(tup3)
```

```python
#coverting tup to list
l=[1,2,3,4,5]
x=tuple(l)
print(x)
```

```python
#updating tuple by converting into list
tup=(1,'situ',89,'appu')
x=list(tup)
x[1]=45
tup1=tuple(x)
print(tup1)
```

```python
#deleting tup
tup=(1,'situ',89,'appu')

del tup

```

```python
#replicating tuple
tup=(1,2,3)
x=tup*3
print(x)
```

```python
#printing sets
a={1,1,2,2,3,4,3,4,5,7,6,7,5,6,8,9}
print(a)
```

```python
#adding elements
a={1,2,3,4,5,6}
a.add(10)
print(a)
```

```python
#add set elements in another set
a={1,2,3,4}
b={5,6,7,8}
a.update(b)
print(a)
```

```python
#Remove element from set
a={1,2,3,4,5}
a.remove(4)
print(a)
a.pop()
print(a)
a.clear()
print(a)
```

```python
#builtin functions on sets
set1={1,2,3,4}
set2={2,5,6,7,8}
union_set=set1.union(set2)
print(union_set)
inter=set1.intersection(set2)
print(inter)
diff=set1.difference(set2)
print(diff)
sub=set1.issubset(set2)
print(sub)
super=set1.issuperset(set2)
print(super)
```
